```python
import os
import pandas as pd
from os import listdir
from os import path
from os.path import isfile, join
import numpy as np
import string 
from bs4 import BeautifulSoup as bs
import string
import json
import re
import copy
from collections import Counter
```

```python
def create_character_cast(play_soup):
    dramatic_characters = play_soup.find_all('role')
    character_dict = {}
    for character_tag in dramatic_characters:
        tag = str(character_tag)
        xml_id = tag[tag.find('id='):tag.find(' sex')].replace('\"', '').split('=')[-1]
        #in case there is a collective number 
        collective_number = character_tag.find_all('collective_number')
        if len(collective_number) != 0:
            collect_number = int(collective_number[0].get_text())
        else:
            collect_number = None
        character_dict[character_tag.get_text()] = {"alternative_names": xml_id,
                                                   "collective_number": collect_number}   
    return character_dict
```

```python
 all_files = [f for f in listdir('TEI_files/') if f.count('.xml')>0]
```

```python
for file_name in all_files:
    with open('TEI_files/'+file_name, 'r') as file:
        print(file_name)
        soup = bs(file, 'lxml')
        cast = create_character_cast(soup)
        speakers = [sp['who'] for sp in soup.find_all('sp')]
        print(set(speakers).difference(set([name['alternative_names'] for name in cast.values()])))
```
